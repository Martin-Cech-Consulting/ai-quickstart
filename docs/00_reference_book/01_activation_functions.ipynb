# Activation Functions

```python
from torch import nn, torch
from matplotlib import pyplot as plt
import numpy as np

# Define a plot function
def plot_fn(title, x, y):
    axs = plt.subplot()
    axs.plot(x, y, color="orange")
    axs.set_title(title, pad= 15)
    axs.axis((np.min(x),np.max(x),np.min(y),np.max(y)))
    axs.spines['bottom'].set_position(("data",0))
    axs.spines['left'].set_position(("data",0))
    axs.spines['top'].set_visible(False)
    axs.spines['right'].set_visible(False)
    plt.xticks([-1,1])
    y_tick_max = np.round(y[-1])
    y_tick_min = 0 if y[0] > 0 else y[0]
    plt.yticks([y_tick_min, y_tick_max])
    plt.show()

# Define Input Data
in_data = np.arange(-5,5.1,0.1)
input = torch.from_numpy(in_data)
```

## 01 Rectified Linear Unit (ReLU)

$f(x) = max(0,x)$

```python
plot_fn('ReLU', in_data, nn.ReLU()(input).numpy())
```

## 02 Sigmoid Function (Logistic Function)

$\sigma(x) = \frac{1}{1+e^{-x}}$

```python
plot_fn('Sigmoid', in_data, nn.Sigmoid()(input).numpy())
```

## 03 Softmax Function

$f(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n}{e^{x_j}}}$

```python
plot_fn('Softmax', in_data, nn.Softmax(dim=0)(input).numpy())
```

## 04 Tangens Hyperbolicus (tanh)

$tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

```python
plot_fn('tanh', in_data, nn.Tanh()(input).numpy())
```
